# Read Files

```python
amino_acid = ["A","R","N","D", "B","C","Q","E","G","H","I","L","K","M","F","P","S","T","W","Y","V"]
```

```python
train = open("pa4train.txt")
train = [l.strip().split() for l in train]
train = [["".join([w if w in amino_acid else "X" for w in x])]+[int(y)] for x,y in train]
```

```python
test = open("pa4test.txt")
test = [l.strip().split() for l in test]
test = [["".join([w if w in amino_acid else "X" for w in x])]+[int(y)] for x,y in test]
```

# Functions

```python
def kernel(data, length):
    d = []
    for each in data:
        d_each = dict()
        a = each[0]
        b = each[1]
        for i in range(len(a) - length + 1):
            if not a[i:i+length] in d_each:
                d_each[a[i:i+length]] = 0
            d_each[a[i:i+length]] += 1
        d = d + [[d_each] + [b]]
    return d
```

```python
def modified_kernel(data, length):
    d = []
    for each in data:
        d_each = dict()
        a = each[0]
        b = each[1]
        for i in range(len(a) - length + 1):
            d_each[a[i:i+length]] = 1
        d = d + [[d_each] + [b]]
    return d
```

```python
def modified_add(x, y):
    d = dict()
    for i in x:
        d[i] = x[i]
    for i in y:
        if not i in d:
            d[i] = 0
        d[i] = d[i] + y[i]
    return d
```

```python
def modified_dot(x, y):
    result = 0
    for i in x:
        if i in y:
            result = result + x[i] * y[i]
    return result
```

```python
def modified_mul(x,y):
    d = dict()
    for i in y:
        d[i] = x * y[i]
    return d
```

```python
def modified_perception(data):
    d = dict()
    for each in data:
        x = each[0]
        y = each[1]
        thresh = y * modified_dot(d,x)
        if thresh <= 0:
            d = modified_add(d, modified_mul(y,x))
    return d
```

```python
import random
def get_error(data, s, p):
    c = 0
    for i in range(len(data)):
        thresh = modified_dot(s[i][0], p)
        if thresh > 0:
            sign = 1
        elif thresh < 0:
            sign = -1
        else:
            sign = random.choice([-1, 1])
        if(sign != data[i][-1]):
            c = c + 1
    return c / len(data)
```

# Question 1

```python
print("Errors: ")
print()
for i in range(2, 6, 1):
    s_train = kernel(train, i)
    p = modified_perception(s_train)
    train_error = get_error(train, s_train, p)
    
    s_test = kernel(test, i)
    test_error = get_error(test, s_test, p)
    
    print("p = ", i, ":")
    print("Train Error:",train_error)
    print("Test Error:",test_error)
```
